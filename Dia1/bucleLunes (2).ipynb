```python
while True:
  valor = input ('Introduce un entero [type q to quit]:')
  if valor == 'q':
    break 
  numero = int (valor)
  if numero % 2 == 0:
    print ('Es par, saltamos bucle')
    continue 
  cuadrado = numero * numero 
  print (numero, 'el cuadrado es: ', cuadrado)
```
